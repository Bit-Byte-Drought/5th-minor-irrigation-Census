```python
def subset_blr(loc):
    df = pd.read_csv(loc)
    df['District'] = df['District'].apply(lambda x: np.nan if 'bangalore' not in str(x).lower() else x)
    df = df.dropna()
    return df
```

A function to return instances from only Bangalore.

```python
li = list()
for i in loc:
    li.append(subset_blr(i))
```

```python
for i in range(len(li)):
    li[i] = Clean.clean(li[i])
```
